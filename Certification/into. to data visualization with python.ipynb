# Line plot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(1,15)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color="g",linestyle=":",linewidth=5)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="b",linestyle=":",linewidth=5)
plt.plot(x,y2,color="r",linestyle="-.",linewidth=4)
plt.title("Two Lines Plot between y1 & y2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color="b",linestyle=":",linewidth=5)

plt.subplot(2,1,2)
plt.plot(x,y2,color="r",linestyle="-.",linewidth=4)

plt.title("Lines Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

# Bar Plot

```python
student={"Ahmad":20,"Awais":37,"Laeeq":33}
```

```python
names=list(student.keys())
values=list(student.values())
plt.title("Performance")
plt.xlabel("Names")
plt.ylabel("Percentage")
plt.bar(names,values)
plt.show()
```

```python
names=list(student.keys())
values=list(student.values())
plt.title("Performance")
plt.xlabel("Percentage")
plt.ylabel("Names")
plt.barh(names,values,color="g")
plt.show()
```

# Scatter plot

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,8,7,6,5,4,3,2,1]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,7,8,3,1,4,6,5,2]
plt.scatter(x,y,marker="*",c="g",s=100)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,7,8,3,1,4,6,5,2]
z=[5,1,4,2,7,3,9,4,5]
plt.scatter(x,y,marker="*",c="g",s=100)
plt.scatter(x,z,marker=".",c="r",s=100)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,7,8,3,1,4,6,5,2]
z=[5,1,4,2,7,3,9,4,5]

plt.subplot(1,2,1)
plt.scatter(x,y,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,z,marker=".",c="r",s=100)
plt.show()
```

# boxplot

```python
one=[1,2,3,8,9]
two=[9,8,3,2,1]
three=[9,8,7,6,5]

data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

# violin plot

```python
one=[1,2,3,8,9]
two=[9,8,3,2,1]
three=[9,8,7,6,5]

data=list([one,two,three])
plt.violinplot(data,showmedians=True)
plt.show()
```

# DoughNut Chart

```python
fruit=['Mango','Apple','Orange','Stobery']
Quantity=[60,55,22,25]
```

```python
plt.pie(Quantity, labels=fruit, radius=2, wedgeprops={'width': 0.9})
plt.show()

plt.pie([100], radius=1, colors=['w'])
plt.show()
```
