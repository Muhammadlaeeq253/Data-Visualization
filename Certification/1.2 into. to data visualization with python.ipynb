# Seaborn library use as sns

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
fmri.shape
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

# Bar plot 

```python
import pandas as pd
sns.set(style="whitegrid")
pokemon=pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",hue="generation",data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,palette="rocket")
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,palette="Blues_d")
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,palette="vlag")
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,color="Orange")
plt.show()
```

# Histogram / Distplot

```python
diamonds=pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.displot(diamonds['price'])
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color="green")
plt.show()
```

```python
sns.distplot(diamonds['price'],color="red")
plt.show()
```

```python
sns.distplot(diamonds['price'],color="purple")
plt.show()
```

```python
sns.distplot(diamonds['price'],color="blue",kde=False,bins=20)
plt.show()
```

```python
sns.distplot(diamonds['price'],color="yellow",vertical=True)
plt.show()
```

```python
sns.distplot(diamonds['price'],color="yellow",kde=False,bins=20,vertical=True)
plt.show()
```

# Joint Plot

```python
iris=pd.read_csv('iris.csv')
iris.head()
```

```python
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,color="olive")
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,color="brown",kind="reg")
plt.show()
```

```python
churn=pd.read_csv("churn.csv")
churn.head()
```

```python
sns.boxplot(x="Churn",y="tenure",data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService",y="MonthlyCharges",data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService",y="MonthlyCharges",data=churn,palette="Set1")
plt.show()
```

```python
sns.boxplot(x="InternetService",y="tenure",data=churn,color="blue")
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn,color="orange",linewidth=3)
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn,order=["One year","Month-to-month","Two year"])
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn,hue="PaymentMethod")
plt.show()
```
